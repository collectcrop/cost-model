```python
import pandas as pd
from io import StringIO


df = pd.read_csv("./result/pgm_disk_multithread.csv")

# fixed_ios = 1014295
fixed_ios = 1000000

df["IOPS"] = fixed_ios / df["avg_walltime_s"]
iops_summary = df.groupby("threads")["IOPS"].mean().reset_index()

print(iops_summary)
```
